```python
import os
from time import time
import multiprocessing
import netCDF4 as nc
import xarray as xr
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas
```

```python
epoch = np.datetime64('1970-01-01T00:00:00')
T = ((12*60+25)*60)/2 


def peak_trough_idx(height,t):
    '''
    indices of the peaks and troughs
    
    '''

    
    
    secs = (t-epoch)/1e9 # dtype is timedelta[s] not timedelta[ns]
    secs = secs.astype('float64')
    
    step = 25
    istep = 20
    N = (secs[-1]-secs[0])/T
    
    if N%1 > 0.4 and N%1 < 0.9:
        
        n = int(N)
            
        idx_troughs = np.zeros((n+1))
        idx_peaks = np.zeros((n+1))
        idx_troughs[:] = np.nan
        idx_peaks[:] = np.nan

        
    else:
    
        idx_troughs = np.zeros((round(N)))
        idx_peaks = np.zeros((round(N)))
        idx_troughs[:] = np.nan
        idx_peaks[:] = np.nan

    
    
    
   

    if np.isnan(height[0]) == True:
        pass



    if height[0]>height[1] or height[0]==height[1] and height[1]>height[2]:

        if N%1 > 0.4 and N%1 < 0.9:
            peak_shape = n
            trough_shape = n+1

            trough = min(height[:istep])
            idx_trough = [x for x, z in enumerate(height[:istep]) if z == trough][0]
            if height[idx_trough] == height[istep-1]:
                trough = min(height[:istep+step])
                idx_trough = [x for x, z in enumerate(height[:istep+step]) if z == trough][0]
            idx_troughs[0] = idx_trough



            for i in range(1,trough_shape):

                peak_slice = height[idx_trough:idx_trough+step]
                peak = max(peak_slice)
                idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                if height[idx_peak] == peak_slice[-1]:
                    peak_slice = height[idx_trough:idx_trough+step+step]
                    peak = max(peak_slice)
                    idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                idx_peaks[i-1] = idx_peak


                trough_slice = height[idx_peak:idx_peak+step]
                trough = min(trough_slice)
                idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                if height[idx_trough] == trough_slice[-1]:
                    trough_slice = height[idx_peak:idx_peak+step+step]
                    trough = min(trough_slice)
                    idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                idx_troughs[i] = idx_trough





        else:
            peak_shape = round(N)
            trough_shape = round(N)

            trough = min(height[:istep])
            idx_trough = [x for x, z in enumerate(height[:istep]) if z == trough][0]
            if height[idx_trough] == height[istep-1]:
                trough = min(height[:istep+step])
                idx_trough = [x for x, z in enumerate(height[:istep+step]) if z == trough][0]
            idx_troughs[0] = idx_trough




            for i in range(1,trough_shape):


                peak_slice = height[idx_trough:idx_trough+step]
                peak = max(peak_slice)
                idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                if height[idx_peak] == peak_slice[-1]:
                    peak_slice = height[idx_trough:idx_trough+step+step]
                    peak = max(peak_slice)
                    idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                idx_peaks[i-1] = idx_peak

                trough_slice = height[idx_peak:idx_peak+step]
                trough = min(trough_slice)
                idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                if height[idx_trough] == trough_slice[-1]:
                    trough_slice = height[idx_peak:idx_peak+step+step]
                    trough = min(trough_slice)
                    idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                idx_troughs[i] = idx_trough




            peak_slice = height[idx_trough:idx_trough+step]
            peak = max(peak_slice)
            idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
            if height[idx_peak] == peak_slice[-1]:
                peak_slice = height[idx_trough:idx_trough+step+step]
                peak = max(peak_slice)
                idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
            idx_peaks[-1] = idx_peak





    elif height[0]<height[1] or height[0]==height[1] and height[1]<height[2]:

        if N%1 > 0.4 and N%1 < 0.9:
            peak_shape = n+1
            trough_shape = n

            peak = max(height[:istep])
            idx_peak = [x for x, z in enumerate(height[:istep]) if z == peak][0]
            if height[idx_peak] == height[istep-1]:
                peak = max(height[:istep+step])
                idx_peak = [x for x, z in enumerate(height[:istep+step]) if z == peak][0]
            idx_peaks[0] = idx_peak




            for i in range(1,peak_shape):

                trough_slice = height[idx_peak:idx_peak+step]
                trough = min(trough_slice)
                idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                if height[idx_trough] == trough_slice[-1]:
                    trough_slice = height[idx_peak:idx_peak+step+step]
                    trough = min(trough_slice)
                    idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                idx_troughs[i-1] = idx_trough


                peak_slice = height[idx_trough:idx_trough+step]
                peak = max(peak_slice)
                idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                if height[idx_peak] == peak_slice[-1]:
                    peak_slice = height[idx_trough:idx_trough+step+step]
                    peak = max(peak_slice)
                    idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                idx_peaks[i] = idx_peak






        else:
            peak_shape = round(N)
            trough_shape = round(N)

            peak = max(height[:istep])
            idx_peak = [x for x, z in enumerate(height[:istep]) if z == peak][0]
            if height[idx_peak] == height[istep-1]:
                peak = max(height[:istep+step])
                idx_peak = [x for x, z in enumerate(height[:istep+step]) if z == peak][0]
            idx_peaks[0] = idx_peak


            for i in range(1,peak_shape):


                trough_slice = height[idx_peak:idx_peak+step]
                trough = min(trough_slice)
                idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                if height[idx_trough] == trough_slice[-1]:
                    trough_slice = height[idx_peak:idx_peak+step+step]
                    trough = min(trough_slice)
                    idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
                idx_troughs[i-1] = idx_trough


                peak_slice = height[idx_trough:idx_trough+step]
                peak = max(peak_slice)
                idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                if height[idx_peak] == peak_slice[-1]:
                    peak_slice = height[idx_trough:idx_trough+step+step]
                    peak = max(peak_slice)
                    idx_peak = [x for x, z in enumerate(peak_slice) if z == peak][0] + idx_trough
                idx_peaks[i] = idx_peak



            trough_slice = height[idx_peak:idx_peak+step]
            trough = min(trough_slice)
            idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
            if height[idx_trough] == trough_slice[-1]:
                trough_slice = height[idx_peak:idx_peak+step+step]
                trough = min(trough_slice)
                idx_trough = [x for x, z in enumerate(trough_slice) if z == trough][0] + idx_peak
            idx_troughs[-1] = idx_trough



           
                
               

                    

    return idx_peaks,idx_troughs
```

```python
ds_range_pow = xr.open_dataset('data/processed/range-power.nc')

latitude = ds_range_pow.lat.values
longitude = ds_range_pow.lon.values
t = ds_range_pow.time.values



r_power = ds_range_pow.power.values
```

```python
modified_sol = pandas.read_csv('data/processed/ga-21-sol-modified.csv')
```

```python
sol_power = 0
plt.figure()
for i in range(20):
    curve = r_power[:,modified_sol.iloc[i]['lat_index'],modified_sol.iloc[i]['lon_index']]
    sol_power += curve
    plt.plot(t,curve,label=f'{modified_sol.iloc[i,0]}')
    
    
plt.figure()
plt.plot(t,sol_power)
    
    
sol_mean = np.mean(sol_power)
rmsd = np.sqrt(np.mean((sol_power-sol_mean)**2))
print((sol_mean**2)/rmsd)
```
